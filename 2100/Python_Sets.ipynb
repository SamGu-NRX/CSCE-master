### Question 3

```python
### Initial Set-up

A = {1, 3, 5, 7}
B = {2, 4, 6, 8}
C = {1, 5}
```

```python
### Method for cross-product of 2 sets ###
""" Python sets do not have an inbuilt method that returns the cross-product of 2 sets,
  so I am writing a custom one """

def cross_product(X: set, Y: set) -> set:
  cross_set = set()
  for x in X:
    for y in Y:
      cross_set.add((x, y))
  return cross_set
```

```python
""" The itertools 'product' functions does not return a set,
  therefore I am using my custom function """

from itertools import product

A = {1}
B = {2}

cartesian_1 = product(A, B)
cartesian_2 = cross_product(A, B)

print(isinstance(cartesian_1, set))
print(isinstance(cartesian_2, set))
```

```python
# (a)

new_set = cross_product(A, B)
print(new_set)
```

```python
# (b)

new_set = cross_product(C, A)
print(new_set)
```

```python
# (c)

new_set = cross_product(B, C)
print(new_set)
```
