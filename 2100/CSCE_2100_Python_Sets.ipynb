```python
# Question 1
print("Question 1:")

# Define the sets
A = {3, 6, 9}
B = {4, 5, 9, 10}
C = {2, 4, 11, 14}
D = {3, 6, 9}
E = {4, 6, 16}

# a. B ∪ C
union_B_C = B.union(C)
print("a. B ∪ C =", union_B_C)
# Expected: {2, 4, 5, 9, 10, 11, 14}

# b. A ∩ D
intersection_A_D = A.intersection(D)
print("b. A ∩ D =", intersection_A_D)
# Expected: {3, 6, 9}

# c. A ∩ C
intersection_A_C = A.intersection(C)
print("c. A ∩ C =", intersection_A_C)
# Expected: set() (the empty set)

# d. (B ⊕ D) ⊕ E
# In Python, symmetric difference is denoted by ^
symmetric_B_D = B ^ D
result_d = symmetric_B_D ^ E
print("d. (B ⊕ D) ⊕ E =", result_d)
# Expected: {3, 5, 10, 16}

# e. A ∪ D
union_A_D = A.union(D)
print("e. A ∪ D =", union_A_D)
# Expected: {3, 6, 9}

# f. E – D
diff_E_D = E - D
print("f. E – D =", diff_E_D)
# Expected: {4, 16}

# g. D – E
diff_D_E = D - E
print("g. D – E =", diff_D_E)
# Expected: {3, 9}

print("\n" + "="*50 + "\n")

# Question 2
print("Question 2:")

# Define the sets
A = {1, 2, 3, 4}
B = {-1, -2, -3}
C = {-1, 0, 1, 2, 3, 4}
D = {2, 3, 4, 5, 6, 7}

# a. |A ∪ B|
union_A_B = A.union(B)
cardinality_union_A_B = len(union_A_B)
print("a. |A ∪ B| =", cardinality_union_A_B)
# Expected: 7

# b. A ⊕ B
symmetric_A_B = A ^ B
print("b. A ⊕ B =", symmetric_A_B)
# Expected: {-1, -2, -3, 1, 2, 3, 4}

# c. A ⊕ A
symmetric_A_A = A ^ A
print("c. A ⊕ A =", symmetric_A_A)
# Expected: set() (the empty set)

print("\n" + "="*50 + "\n")

# Question 3
print("Question 3:")

# Define the sets
A = {1, 3, 5, 7}
B = {2, 4, 6, 8}
C = {1, 5}

# a. A X B (Cartesian product)
cartesian_A_B = {(a, b) for a in A for b in B}
print("a. A X B =", cartesian_A_B)
# Expected: {(1,2), (1,4), (1,6), (1,8), (3,2), ..., (7,8)}

# b. C X A
cartesian_C_A = {(c, a) for c in C for a in A}
print("b. C X A =", cartesian_C_A)
# Expected: {(1,1), (1,3), (1,5), (1,7), (5,1), (5,3), (5,5), (5,7)}

# c. B X C
cartesian_B_C = {(b, c) for b in B for c in C}
print("c. B X C =", cartesian_B_C)
# Expected: {(2,1), (2,5), (4,1), (4,5), (6,1), (6,5), (8,1), (8,5)}

print("\n" + "="*50 + "\n")

# Question 4
print("Question 4:")

# Define the directed graph as a set of edges
graph_edges = {(1, 2), (1, 3), (1, 4), (3, 4), (3, 5),
               (4, 2), (5, 4), (5, 6), (6, 4)}

print("Directed Graph edges =", graph_edges)

# Optionally, create and display an adjacency list for the graph
from collections import defaultdict

adjacency_list = defaultdict(list)
for u, v in graph_edges:
    adjacency_list[u].append(v)

print("\nAdjacency List Representation:")
for u in sorted(adjacency_list):
    print(f"{u} -> {sorted(adjacency_list[u])}")
```
